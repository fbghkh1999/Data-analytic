```python
addones=glob.glob(addr1)
addtwos=glob.glob(addr2)
addfives=glob.glob(addr5)
ones1=[]
twos=[]
fives=[]
```

# descriptor

```python
def descriptor(img):
    dim=(3,3)
    resize=cv2.resize(img,dim,interpolation=cv2.INTER_LINEAR)
    return resize.flatten()
```

```python
descriptor(ones1[0])
```

# Train

```python
x=[]
y=[]
for i in ones1:
    des=descriptor(i)
    x.append(des)
    y.append(1)
for i in twos:
    des=descriptor(i)
    x.append(des)
    y.append(2)
for i in fives:
    des=descriptor(i)
    x.append(des)
    y.append(5)
```

```python
clf=svm.SVC()
clf.fit(x,y)
```

```python
addrt="/home/f/Desktop/binaii code/tamrin7ghayor/HodaSmallSubset/*.png"
datatest=glob.glob(addrt)
datatest
```

```python
tests=[]
predict=[]
for i in range(len(datatest)):
    img=np.array(Image.open(datatest[i]))
    tests.append(img)
for j in tests:
    des=descriptor(j)
    res=clf.predict([des])
    predict.append(res)
```
